```python
#loading the hand written data sets
import pandas as pd
from sklearn.datasets import load_digits
```

```python
# 1. Create a Pandas DataFrame from the load_digits dataset
# Load the digits dataset
digits = load_digits()

# Create a Pandas DataFrame
data = pd.DataFrame(digits.data, columns=[f'pixel_{i}' for i in range(digits.data.shape[1])])

# Add the target column to the DataFrame
data['target'] = digits.target

data
```

```python
# 2. Perform EDA on the data. Show the shape, data types, first five lines.
# Ensure there are no missing values.
# 2.1 shape
 #Display the shape of the DataFrame
print("Shape of the DataFrame:")
print(data.shape)
```

```python
#2.2 data types
# Display the data types of each column
print("Data Types:")
print(data.dtypes)
```

```python
# 2.3 first five lines.
# Display the first five lines of the DataFrame
print("\nFirst five lines of the DataFrame:")
print(data.head())
```

```python
#2.4 Show all columns. Are the names useful? What would useful column names be in this case?
# Display all columns
print("All Columns:")
print(data.columns)
```

```python
# 2.5 ensure no null data.
# Check for missing values
print("Missing Values:")
print(data.isnull().sum())
```

```python
#3. perform standardization by importing necessary libraries
from sklearn.preprocessing import StandardScaler
# Extract features (pixel values) for standardization
features = data.drop('target', axis=1)

#3.1 Import and apply the StandardScaler. Save the scaled dataframe to "df_standardized"
scaler = StandardScaler()
df_standardized = scaler.fit_transform(features)

# Create a new DataFrame with standardized features
df_standardized = pd.DataFrame(df_standardized, columns=features.columns)
```

```python
# 3.2 Display (part of) the original and standardized dataframes. 
print("Original DataFrame:")
print(data.head())

print("\nStandardized DataFrame:")
print(df_standardized.head())
```

```python
# 3.3 Display the scaled features as a dataframe
print("\nScaled Features as a DataFrame:")
print(df_standardized.head())
```

```python
# 4. show the correlation between columns
corr_matrix = data.corr()
print("Correlation Between Columns:")
print(corr_matrix)
```

```python
# 5. perform PCA with no reducing(include all columns) 
#5.1 PCA goes here.
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA()
pca_output = pca.fit_transform(features)
```

```python
# 5.2 Display the first 10 lines of the PCA output.
print("First 10 lines of the PCA output:")
print(pd.DataFrame(pca_output).head(10))
```

```python
# 5.3 Draw a bar chart showing the explained_variance of all PC's 
explained_variance_ratio = pca.explained_variance_ratio_
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance of Principal Components')
plt.show()
```

```python
# 5.4 Display the explained_variance_ for the model.
print("\nExplained Variance for the model:")
print(pca.explained_variance_)
```

```python
# 6. Display scatter plots comparing the first, second, and third most important 
# principal components in the PCA model
plt.figure(figsize=(16, 8))

# Scatter plot for the first two principal components
plt.subplot(1, 3, 1)
plt.scatter(pca_output[:, 0], pca_output[:, 1], c=data['target'], cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot: PC1 vs PC2')

# Scatter plot for the first and third principal components
plt.subplot(1, 3, 2)
plt.scatter(pca_output[:, 0], pca_output[:, 2], c=data['target'], cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.title('Scatter Plot: PC1 vs PC3')

# Scatter plot for the second and third principal components
plt.subplot(1, 3, 3)
plt.scatter(pca_output[:, 1], pca_output[:, 2], c=data['target'], cmap='viridis')
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('Scatter Plot: PC2 vs PC3')

plt.show()
```

```python
# 7. Perform a PCA using only the top 10 components (n_components=10)
# 7.1 Code goes here.
pca_top10 = PCA(n_components=10)
pca_top10_result = pca_top10.fit_transform(features)
```

```python
# 7.2 Display the first 10 lines of the PCA
print("First 10 lines of PCA output with n_components=10:")
print(pd.DataFrame(pca_top10_result).head(10))
```

```python
# 7.3 Draw a bar chart showing the explained_variance of the top 10 PC's 
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca_top10.explained_variance_) + 1), pca_top10.explained_variance_)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.title('Explained Variance of the Top 10 Principal Components')
plt.show()
```

```python
# 7.4 Display the explained_variance_ for the model.
print("Explained Variance of Top 10 Principal Components:")
print(pca_top10.explained_variance_)
```

```python
# 8. Display scatter plots comparing the first, second, and third most important 
# principal components in the smaller PCA model
plt.figure(figsize=(16, 8))

# First vs. Second PC
plt.subplot(1, 3, 1)
plt.scatter(pca_top10_result[:, 0], pca_top10_result[:, 1], c=data['target'], cmap='viridis', edgecolor='k')
plt.title('Scatter Plot: First vs. Second PC (Top 10)')

# First vs. Third PC
plt.subplot(1, 3, 2)
plt.scatter(pca_top10_result[:, 0], pca_top10_result[:, 2], c=data['target'], cmap='viridis', edgecolor='k')
plt.title('Scatter Plot: First vs. Third PC (Top 10)')

# Second vs. Third PC
plt.subplot(1, 3, 3)
plt.scatter(pca_top10_result[:, 1], pca_top10_result[:, 2], c=data['target'], cmap='viridis', edgecolor='k')
plt.title('Scatter Plot: Second vs. Third PC (Top 10)')

plt.show()
```
